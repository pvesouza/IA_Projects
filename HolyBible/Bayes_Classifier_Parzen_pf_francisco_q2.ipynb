# Separando os conjuntos entre teste e treinamento


```python
from pandas.core.common import random_state
from sklearn.model_selection import train_test_split

data = joinDataset.to_numpy()
nRow, nCol =  data.shape
print(nRow)
print(nCol)

Y = data[:,-1]
X = data[:, 0 : nCol - 1]
#Separando entre dados de treinamento e teste


X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8, random_state = 10)

print(np.min(X_train[:,3]))
```

#Separando um conjunto de validação a partir do conjunto de teste

```python
print(pd.DataFrame(Y_train).value_counts())
print(pd.DataFrame(X_train).shape)
```

```python
# Lista com vários h para testar qual o melhor classificador
H = []
for i in range(1, 10):
  i = i * 0.1
  H.append(i)
```

#Implementando o modelo Nayve Bayes com Janela de Parzen

```python
from numpy.ma.core import reshape
from scipy.stats import multivariate_normal
from sklearn.neighbors import KernelDensity

classes = joinDataset['F11'].unique()

#Lista de lista com todas as Probabilidades para cada classe
DataList = []
for h in H:
  #Matriz para armazenar as probabilidades de cada classe
  P = pd.DataFrame(data=np.zeros((X_train.shape[0], len(classes))), columns=classes)
  #Fração de elementos de cada classes
  Pc = np.zeros(len(classes))
  #For para iterar sobre cada classe i
  for i in range(0, len(classes)):
    elements = tuple(np.where(Y_train == classes[i]))
    Pc[i] = len(elements) / len(Y_train)                                  #Probabilidades a priori
    Z = X_train[elements,:][0]
    kernel = KernelDensity(kernel='gaussian', bandwidth=h).fit(Z)
    for j in range(0, X_test.shape[0]):
      x = X_test[j,:]
      x = x.reshape((1, len(x)))
      pj = np.exp(kernel.score_samples(x))
      P[classes[i]][j] = Pc[i] * pj
  DataList.append(P)
```

```python
print(DataList)
```

#Predição

```python
from numpy.lib.function_base import append
Y_pred = []
y_pred = []

for j in range(len(H)):
  y_pred.clear()
  for i in range(0, Y_test.shape[0]):
    c = np.argmax(np.array(DataList[j].iloc[[i]]))
    y_pred.append(classes[c])
  #Juntando na lista todos os y preditos
  Y_pred.append(np.array(y_pred, dtype=str))
```

```python
#Avaliando o preditor
from sklearn.metrics import accuracy_score
scores = []
print(Y_pred[2].shape)
print(Y_test.shape)

for i, h in enumerate(H):
  scores.append([accuracy_score(Y_pred[i], Y_test), H[i]])

scores = np.array(scores)
max = 0
max_h = 0

for i in range(0, len(scores)):
  if max < scores[i][0]:
    max = scores[i][0]
    max_h = scores[i][1]

print("Score Maximo = ", max * 100)
print("h = ", max_h)
```
